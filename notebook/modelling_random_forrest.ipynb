## Confusion matrix

```python
modelling.confusion_matrix(y_test,predictions)
```

## Interpretation

True Positives (TP): 2299 - The model correctly predicted 2299 instances of the positive class (default).

True Negatives (TN): 26399 - The model correctly predicted 26399 instances of the negative class (non-default).

False Positives (FP): 0 - The model incorrectly predicted 0 instances as positive (default) when they were actually negative (non-default).

False Negatives (FN): 0 - The model incorrectly predicted 0 instances as negative (non-default) when they were actually positive (default).

Metrics:

Accuracy: (TP + TN) / (TP + TN + FP + FN) = (2299 + 26399) / (2299 + 26399 + 0 + 0) = 1.0

Precision: TP / (TP + FP) = 2299 / (2299 + 0) = 1.0

Recall: TP / (TP + FN) = 2299 / (2299 + 0) = 1.0

F1-score: 2 * (precision * recall) / (precision + recall) = 1.0

**Insights:**

**Perfect Classification:** The confusion matrix shows that the model correctly classified all instances, resulting in perfect accuracy, precision, recall, and F1-score.

**No Errors:** The model did not make any false positive or false negative predictions, indicating that it is able to perfectly distinguish between the two classes.
